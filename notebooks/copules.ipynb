```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(42)

n_samples = 1000
rho = 0.5
mean = np.array([0,0])
cov = np.array([
            [1, rho],
            [rho,1]
        ])
X = np.random.multivariate_normal(mean, cov, n_samples)
U = stats.norm.cdf(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

```python
Y = np.column_stack([
	stats.beta.ppf(U[:, 0], 1, 2),
	stats.gamma.ppf(U[:, 1], 3, scale=2)
])
```

```python
plt.scatter(Y[:,0],Y[:,1])
```
